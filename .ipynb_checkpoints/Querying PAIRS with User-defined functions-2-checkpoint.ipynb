```python
import folium
import branca
import branca.colormap as cm
```

```python
latmin
```

```python
# helper function to calculate colors
def colorize(array, cmax, cmin=0, cmap='rainbow'):
    arr_norm = (array-cmin)/(np.nanmax(array)-cmin)
    colormap = plt.cm.get_cmap(cmap)
    colors = [plt.cm.get_cmap(cmap)(x) for x in np.linspace(0,1,11)]
    return colormap(arr_norm), colors

# helper function to avoid folium smoothing of large grid cells
def upsample_raster(raster, upsample):
    data = raster.repeat(upsample, axis = 0).repeat(upsample, axis = 1)
    return data
```

```python
data_plot = upsample_raster(data, 5)

m = folium.Map(location = [(latmin+latmax)/2.0, (lonmin+lonmax)/2.0], zoom_start=7,)
image_bounds = [[bbox['minLatitude'], bbox['minLongitude']], [bbox['maxLatitude'], bbox['maxLongitude']]]
cmax = np.nanmax(data_plot)
cmin = np.nanmin(data_plot)

image = np.ma.masked_where(np.isnan(data_plot), data_plot)
colormap = 'inferno'

image_color, colors = colorize(image, cmax, cmin, colormap)

folium.raster_layers.ImageOverlay(image_color, bounds = image_bounds, \
                                  pixelated=True, opacity=0.7).add_to(m)
m.fit_bounds(image_bounds)

m.add_child(cm.LinearColormap(colors=colors, vmin=cmin,vmax=cmax))

folium.LayerControl().add_to(m)

m
```

## PAIRS UDF to query monthly climatology

### Helper function to query PAIRS and create climatological average for the specified month, over the year range from year_start to year_end

```python
def get_PAIRS_monthly_clim(layer_id, month, year_start, year_end, spatial):        
    layerList = list()
    # Query PAIRS for mean monthly values for each year
    for year in range(year_start, year_end + 1):        
        if(month!=12):
            end_str = datetime(year, month+1, 1).isoformat()
        else:
            end_str = datetime(year+1, 1, 1).isoformat()
            
        layer_year = {
                 "alias" : "val" + str(year), "type" : "raster", "id" : layer_id,
                 "aggregation" : "Mean",
                 "temporal" : { "intervals" : [{
                         "start" : datetime(year, month, 1).isoformat(),
                         "end" : end_str }]},
                 "output" : False}
        # append list
        layerList.append(layer_year)

    # Create UDF for average over all years
    average_udf = dict()
    average_udf["alias"] = "avg"

    # Create string expression for average to pass to PAIRS
    str_avg = "("
    for year in range(year_start, year_end):
        str_avg += '$val' + str(year) + '+'
    str_avg += '$val' + str(year_end)

    # divide by number of years
    str_avg += ')/(' + str(year_end) + '-' + str(year_start) + '+1)'

    average_udf["expression"] = str_avg

    # append
    layerList.append(average_udf)

    query_json = {"layers" : layerList,
        "spatial" : spatial,
        "temporal" : {"intervals" : [ {"snapshot" : datetime(year_end, month, 1).isoformat()}]}
    }
        
    return query_json
```

## Example to query PAIRS for monthly climatology for all months of the year

```python
str(0).zfill(2)
```

```python
def get_PAIRS_allmonths_clim(layer_id, year_start, year_end, spatial):        
    layerList = list()
    for month in range(1, 13):
        monthstr = str(month).zfill(2)
        for year in range(year_start, year_end + 1):        
            if(month!=12):
                end_str = datetime(year, month+1, 1).isoformat()
            else:
                end_str = datetime(year+1, 1, 1).isoformat()
            
            layer_year = {
                 "alias" : "val" + str(year) + monthstr, "type" : "raster", "id" : layer_id,
                 "aggregation" : "Mean",
                 "temporal" : { "intervals" : [{
                         "start" : datetime(year, month, 1).isoformat(),
                         "end" : end_str }]},
                 "output" : False}
            # append list
            layerList.append(layer_year)

        # Create UDF for average
        
        average_udf = dict()
        average_udf["alias"] = "avg" + monthstr
        

        # Create expression for average 
        str_avg = "("
        for year in range(year_start, year_end):
            str_avg += '$val' + str(year) + monthstr + '+'
        str_avg += '$val' + str(year_end) + monthstr

        # divide by number of years
        str_avg += ')/(' + str(year_end) + '-' + str(year_start) + '+1)'

        average_udf["expression"] = str_avg

        # append
        layerList.append(average_udf)

    query_json = {"layers" : layerList,
            "spatial" : spatial,
            "temporal" : {"intervals" : [ {"snapshot" : datetime(year_end, month, 1).isoformat()}]}
    }
        
    return query_json
```

```python
# polygon
spatial = {
        "type" : "poly",
        "polygon": {"wkt": wkt_string}
}
    
json = get_PAIRS_allmonths_clim(layer_id, 2001, 2020, spatial)
```

```python
query3 = paw.PAIRSQuery(json, PAIRS_SERVER,  auth=OAUTH, authType='api-key') 
query3.submit()
```

### Plot climatology for a specified month

```python
def get_clim_layer_id(metadata, month):
    layer_ids = list(pd.DataFrame(metadata).transpose().index)
    match_id = [x for x in layer_ids if 'avg'+str(month).zfill(2) in x][0]
    return match_id

def get_clim_layer(query, month):
    layer_id = get_clim_layer_id(query.metadata, month)
    return query.data[layer_id]
    
def get_clim_range(query):
    retmax = -999999.9
    retmin = 9999999.9
    for month in range(1, 13):
        arr = get_clim_layer(query, month)
        minval = np.nanmin(arr)
        maxval = np.nanmax(arr)
        if minval < retmin:
            retmin = minval
        if maxval > retmax:
            retmax = maxval
    return retmin, retmax
```

```python
month_sel = 9
data_plot = get_clim_layer(query3, month_sel)
data_plot = upsample_raster(data_plot, 5)
# Convert to deg C
data_plot = data_plot - 273.15
m = folium.Map(location = [(latmin+latmax)/2.0, (lonmin+lonmax)/2.0], zoom_start=7,)
image_bounds = [[bbox['minLatitude'], bbox['minLongitude']], [bbox['maxLatitude'], bbox['maxLongitude']]]
cmax = np.nanmax(data_plot)
cmin = np.nanmin(data_plot)
print(cmax)

image = np.ma.masked_where(np.isnan(data_plot), data_plot)
colormap = 'inferno'

image_color, colors = colorize(image, cmax, cmin, colormap)

folium.raster_layers.ImageOverlay(image_color, bounds = image_bounds, \
                                  pixelated=True, opacity=0.7).add_to(m)
m.fit_bounds(image_bounds)

m.add_child(cm.LinearColormap(colors=colors, vmin=cmin,vmax=cmax))

folium.LayerControl().add_to(m)
m
```

## Panel plot for multiple months

```python
fig, axs = plt.subplots(4, 3, True, True, figsize = (16, 16))
monthnames=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
           'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
ix=0
iy=0
vmin, vmax = get_clim_range(query3)
vmin-=273.15
vmax-=273.15
print(vmin, vmax)
for i in range(12):
    data_plot = get_clim_layer(query3, i+1)
    data_plot = data_plot - 273.15
    plot = axs[ix, iy].imshow(data_plot, extent=extent, cmap="bwr", vmin=vmin, vmax=vmax)
    axs[ix,iy].set_title('Mean temperature for ' + monthnames[i])
    fig.colorbar(plot, ax=axs[ix,iy])
    i+=1
    iy+=1
    if iy>2:
        ix+=1
        iy=0
plt.show()
```
