## Construct the PAIRS query for monthly climatology

### This helper function creates the json we need to pass to PAIRS, with a user-defined function to create the climatology

```python
def get_PAIRS_allmonths_clim(layer_ids, year_start, year_end, spatial):        
    layerList = list()
    # loop over the list of PAIRS layers
    for layer_id in layer_ids:
        for month in range(1, 13):
            monthstr = str(month).zfill(2)
            for year in range(year_start, year_end + 1):        
                if(month!=12):
                    end_str = datetime(year, month+1, 1).isoformat()
                else:
                    end_str = datetime(year+1, 1, 1).isoformat()
                # query for monthly average for specific year
                # output=False as we do not return the values, only the average (see below)
                layer_year = {
                 "alias" : "val" + str(layer_id) + "_" + str(year) + monthstr, 
                 "type" : "raster", 
                 "id" : layer_id,
                 "aggregation" : "Mean",
                 "temporal" : { "intervals" : [{
                         "start" : datetime(year, month, 1).isoformat(),
                         "end" : end_str }]},
                 "output" : False}
                # append list
                layerList.append(layer_year)

            # Create UDF for average over years for this PAIRS layer
            average_udf = dict()
            average_udf["alias"] = "avg" + "_" + str(layer_id) + monthstr
        
            # Create expression for average 
            str_avg = "("
            for year in range(year_start, year_end):
                str_avg += '$val' + str(layer_id) + "_" + str(year) + monthstr + '+'
            str_avg += '$val' + str(layer_id) + "_" + str(year_end) + monthstr

            # divide by number of years
            str_avg += ')/(' + str(year_end) + '-' + str(year_start) + '+1)'

            average_udf["expression"] = str_avg

            # append
            layerList.append(average_udf)

    query_json = {"layers" : layerList,
                "spatial" : spatial,
                "temporal" : {"intervals" : [ {"snapshot" : datetime(year_end, month, 1).isoformat()}]}
    }
        
    return query_json
```

### This function calculates the daily mininum, then averages for all days in the same month over all years

```python
def get_PAIRS_clim_dailymin(layer_ids, year_start, year_end, spatial):        
    layerList = list()
    # loop over the list of PAIRS layers
    for pairs_id in layer_ids:
        for month in range(1, 13):
            monthstr = str(month).zfill(2)
            for year in range(year_start, year_end + 1): 
                day_end = calendar.monthrange(year, month)[1]
                # For each day in the month, find the minimum value
                for day in range(1, day_end+1):
                    daystr = str(day).zfill(2)
                    layer_year = {
                     "alias" : "val" + str(pairs_id) + "_" + str(year) + monthstr + daystr, 
                     "type" : "raster", 
                     "id" : pairs_id,
                     "aggregation" : "Min",
                     "temporal" : { "intervals" : [{
                         "start" : datetime(year, month, day, 0).isoformat(),
                         "end" : datetime(year, month, day, 23, 59, 59).isoformat() }]},
                     "output" : False}
                    # append list
                    layerList.append(layer_year)

            # Create UDF for average over years for this PAIRS layer and month
            average_udf = dict()
            average_udf["alias"] = "avg" + "_" + str(pairs_id) + monthstr
        
            # Create expression for average 
            # Use the alias names from above - for all days and years, for this month
            ndays=0
            str_avg = "("
            for year in range(year_start, year_end+1):
                day_end = calendar.monthrange(year, month)[1]
                ndays+=day_end
                for day in range(1, day_end):
                    daystr = str(day).zfill(2)
                    str_avg += '$val' + str(pairs_id) + "_" + str(year) + monthstr + daystr + '+'
                str_avg += '$val' + str(pairs_id) + "_" + str(year_end) + monthstr + daystr

            # divide by number of years and days
            str_avg += ')/(' + str(ndays) + ')'

            average_udf["expression"] = str_avg

            # append
            layerList.append(average_udf)
            
    query_json = {"layers" : layerList,
                "spatial" : spatial,
                "temporal" : {"intervals" : [ {"snapshot" : datetime(year_end, month, 1).isoformat()}]}
    }
        
    return query_json
```

# Plot the climate data 

## Helper functions to get data from queries returned by PAIRS

```python
def get_clim_layer_id(pairs_layer, metadata, month):
    layer_ids = list(pd.DataFrame(metadata).transpose().index)
    match_id = [x for x in layer_ids if 'avg' + "_" + str(pairs_layer) + str(month).zfill(2) in x][0]
    return match_id

def get_clim_layer(query, pairs_layer, month):
    layer_id = get_clim_layer_id(pairs_layer, query.metadata, month)
    return query.data[layer_id]
    
def get_meta_data(query, pairs_layer, month):
    layer_id = get_clim_layer_id(pairs_layer, query.metadata, month)
    return query.metadata[layer_id]
    
def get_clim_range(query, pairs_layer):
    retmax = -999999.9
    retmin = 9999999.9
    for month in range(1, 13):
        arr = get_clim_layer(query, pairs_layer, month)
        minval = np.nanmin(arr)
        maxval = np.nanmax(arr)
        if minval < retmin:
            retmin = minval
        if maxval > retmax:
            retmax = maxval
    return retmin, retmax
```

### Create a panel plot for temperature

```python
query = query1

id_string = pd.DataFrame(query.metadata).transpose().index[0]
bbox = query.metadata[id_string]['details']['boundingBox']
extent = [bbox['minLongitude'], bbox['maxLongitude'], bbox['minLatitude'], bbox['maxLatitude']]

layer_str = 'ERA5 temperature'
layer_id = pairs_dict[layer_str]
fig, axs = plt.subplots(4, 3, True, True, figsize = (16, 16))
monthnames=['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
           'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
ix=0
iy=0
vmin, vmax = get_clim_range(query, layer_id)
vmin-=273.15
vmax-=273.15
print(vmin, vmax)
for i in range(12):
    data_plot = get_clim_layer(query, layer_id, i+1)
    data_plot = data_plot - 273.15
    plot = axs[ix, iy].imshow(data_plot, extent=extent, cmap="bwr", vmin=vmin, vmax=vmax)
    axs[ix,iy].set_title('Mean ' + layer_str + ' for ' + monthnames[i])
    fig.colorbar(plot, ax=axs[ix,iy])
    i+=1
    iy+=1
    if iy>2:
        ix+=1
        iy=0
plt.show()
```

### Local version

```python
def save_year(query, layer_str, years_str, agg_str):
    layer_id = pairs_dict[layer_str]
    for i in range(12):
        data = get_clim_layer(query, layer_id, i+1)
        file_name = datadir + 'data' + '_' + layer_str + agg_str + "_" + str(i).zfill(2) + years_str + '.csv'
        pd.DataFrame(data).to_csv(file_name, header=None, index=None)
    
        file_name = datadir + 'metadata' + "_" + layer_str agg_str + + "_" + str(i).zfill(2) + years_str + '.txt'
        metadata = get_meta_data(query, layer_id, i+1)
        with open(file_name, 'w') as file:
            json.dump(metadata['details'], file)
```
